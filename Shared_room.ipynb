```python
fig, ax = plt.subplots(figsize=(100, 50))
cax = ax.matshow(corr_Factors1, cmap="Blues")
fig.colorbar(cax)

# AÃ±adir anotaciones manualmente
for i in range(corr_Factors1.shape[0]):
    for j in range(corr_Factors1.shape[1]):
        ax.text(j, i, f"{corr_Factors1.iloc[i, j]:.2f}", 
               ha="center", va="center", fontsize=20)

plt.xticks(range(len(corr_Factors1.columns)), corr_Factors1.columns, rotation=90, fontsize=12)
plt.yticks(range(len(corr_Factors1.index)), corr_Factors1.index, fontsize=12)
plt.show()
```

```python
#Declaramos las variables dependientes e independientes para la regresiÃ³n lineal
Vars_Indep= df[['host_total_listings_count', 'calculated_host_listings_count','calculated_host_listings_count_entire_homes']]
Var_Dep= df['host_listings_count']
```

```python
#Se define model como la funciÃ³n de regresiÃ³n lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
import numpy as np
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)

#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__

#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R DeterminaciÃ³n
model.score(Vars_Indep,Var_Dep)

#Corroboramos cual es el coeficiente de DeterminaciÃ³n de nuestro modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)

print("El coeficiente de determinaciÃ³nes: ", coef_Deter)

#Corroboramos cual es el coeficiente de CorrelaciÃ³n de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
print("El coeficiente de correlaciÃ³n es:", coef_Correl)
```

Esto significa que es muy alto ambos coeficientes
