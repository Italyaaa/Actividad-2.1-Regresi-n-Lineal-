***bathrooms_text***

```python
df['bathrooms_text'].unique()
```

```python
# Función para reemplazar los valores
def convertir_baños(valor):
    if pd.isna(valor):  # Si el valor es NaN
        return 0
    if 'shared' in valor.lower():  # Si incluye "shared"
        # Obtener el número inicial antes de "shared"
        try:
            num = float(valor.split()[0])
        except ValueError:  # Si no hay número explícito
            num = 0  # Ejemplo: 'Shared half-bath' -> 0
        return num + 0.05
    elif 'half-bath' in valor.lower():  # Si incluye "half-bath"
        return 0.5
    else:  # Para otros casos como "1 bath", "3 baths", etc.
        try:
            return float(valor.split()[0])
        except ValueError:  # Manejo de errores en caso de formato inesperado
            return 0
```

```python
# Aplicar la función a la columna 'bathrooms_text'
df['bathrooms_text'] = df['bathrooms_text'].apply(convertir_baños)
```
